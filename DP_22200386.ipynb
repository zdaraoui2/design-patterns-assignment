````python
#MVC_Template_01
#2014 May 23  by Steven Lipton http://makeAppPie.com
#Controller initializing MVC -- simplest version possible.
#Additional annotations by R Mather - Oct 2018

# MVC EXTENDED WITH DELETE BUTTON

# Tkinter is a Python binding to the Tk GUI toolkit/library - RM
from tkinter import *
 
# A Model-View-Controller framework for TKinter.
# Model: Data Structure. Controller can send messages to it, and model can respond to message.
# View : User interface elements. Controller can send messages to it. View can call methods from Controller when an event happens.
# Controller: Ties View and Model together. turns UI responses into chages in data.
 


#Controller: Ties View and Model together.
#       --Performs actions based on View events.
#       --Sends messages to Model and View and gets responses
#       --Has Delegates - RM NOTE: essential for communications between view and model via control
 
class MyController():
    def __init__(self,parent):
        self.parent = parent
        # CRITICAL PART OF THE MVC 
        # RM NOTES:
        # MyController passes a reference to itself (self) when it creates MyModel and MyView objects
        # This allows the MyModel and MyView objects to create 'delegates' (see 'vc' = virtual controller?)
        # The vc objects can then call methods on the MyControl object, thereby allowing the View to update the Model via the Controller 
        self.model = MyModel(self)    # initializes the model
        self.view = MyView(self)  #initializes the view
        # Initialize objects in view
        self.view.setEntry_text('Add to Label') # a non cheat way to do MVC with tkinter control variables
        self.view.setLabel_text('Ready')
    
    #event handlers
    def quitButtonPressed(self):
        self.parent.destroy()
    def addButtonPressed(self):
        self.view.setLabel_text(self.view.entry_text.get())
        # CRITICAL PART OF THE MVC 
        # RM NOTES - In one operation:
        #[1] Get the text from the VIEW;
        #[2] Add it to the MODEL by appending it to the list
        self.model.addToList(self.view.entry_text.get())
        
    # *** Adding controller code for remove button - This method is called when the remove button is pressed ***
    # *** It gets the text from the text entry widget and attempts to remove it from the list in the model ***
        self.model.removeFromList(self.view.entry_text.get())
    
    def listChangedDelegate(self):
        #model internally changes and needs to signal a change
        print(self.model.getList())
 
#View : User interface elements.
#       --Controller can send messages to it.
#       --View can call methods from Controller vc when an event happens.
#       --NEVER communicates with Model.
#       --Has setters and getters to communicate with controller
 
class MyView(Frame):
    
    def __init__(self,vc):
        self.frame=Frame()
        self.frame.grid(row = 0,column=0)
        self.vc = vc
        self.entry_text = StringVar()
        self.entry_text.set('nil')
        self.label_text = StringVar()
        self.label_text.set('nil')
        self.loadView()

    def loadView(self):
        quitButton = Button(self.frame,text = 'Quit', command= self.vc.quitButtonPressed).grid(row = 0,column = 0)
        addButton = Button(self.frame,text = "Add", command = self.vc.addButtonPressed).grid(row = 0, column = 1)
         # *** Adding remove button to the view (GUI) which once pressed runs the removedButtonPressed in the controller to remove the item from the list*** 
        removeButton = Button(self.frame,text = "Remove", command = self.vc.removeButtonPressed).grid(row = 0, column = 2)
        entry = Entry(self.frame,textvariable = self.entry_text).grid(row = 1, column = 0, columnspan = 3, sticky = EW)
        label = Label(self.frame,textvariable = self.label_text).grid(row = 2, column = 0, columnspan = 3, sticky = EW)

    def getEntry_text(self):
    #returns a string of the entry text
        return self.entry_text.get()
    def setEntry_text(self,text):
    #sets the entry text given a string
        self.entry_text.set(text)
    def getLabel_text(self):
    #returns a string of the Label text
        return self.label_text.get()
    def setLabel_text(self,text):
    #sets the label text given a string
        self.label_text.set(text)
 
#Model: Data Structure.
#   --Controller can send messages to it, and model can respond to message.
#   --Uses delegates from vc to send messages to the Control of internal change
#   --NEVER communicates with View
#   --Has setters and getters to communicate with Controller
 
class MyModel():
    def __init__(self,vc):
        self.vc = vc
        self.myList = ['duck','duck','goose']
        self.count = 0
#Delegates-- Model would call this on internal change
    def listChanged(self):
        self.vc.listChangedDelegate()
#setters and getters
    def getList(self):
        return self.myList
    def initListWithList(self, aList):
        self.myList = aList
    def addToList(self,item):
        print("returned")
        myList = self.myList
        myList.append(item)
        self.myList=myList
        self.listChanged()
    # *** Adding model code to remove a specific item from the list. 
    # This code first checks if the item is present in the list and then removes it.
    def removeFromList(self, item):
        if (item in self.myList):
            print("returned")
            myList = self.myList
            myList.remove(item)
            # Tells the controller that there has been a change
            self.listChanged()
        else:
            print("The item you are trying to remove is not present in the list")
 
def main():
    # Create a root window from the Tk GUI class
    root = Tk()
    frame = Frame(root,bg='#0555ff' )
    root.title('Hello Penguins')
    app = MyController(root)
    root.mainloop()
 
if __name__ == '__main__':
    main()   
````

````python
import matplotlib.pyplot as plt
import numpy as np


# Nicely abstracted structure by which any model can notify any observer (view) of changes in the model
class Subject(object): #Represents what is being 'observed'

    def __init__(self):
        self._observers = [] # This where references to all the observers are being kept
                             # Note that this is a one-to-many relationship: there will be one subject to be observed by multiple _observers

    def attach(self, observer):
        if observer not in self._observers: #If the observer is not already in the observers list
            self._observers.append(observer) # append the observer to the list

    def detach(self, observer): #Simply remove the observer
        try:
            self._observers.remove(observer)
        except ValueError:
            pass

    def notify(self, modifier=None):
        for observer in self._observers: # For all the observers in the list
            if modifier != observer: # Don't notify the observer who is actually doing the updating 
                observer.update(self) # Alert the observers!
                
# Represents the 'data' for which changes will produce notifications to any registered view/observer objects 
class Model(Subject): # Extends the Subject class

    def __init__(self, name):
        Subject.__init__(self)
        self._name = name # Set the name of the model
        self._labels = ['Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp']
        self._data = [10,8,6,4,2,1]

    @property #Getter that gets the labels
    def labels(self):
        return self._labels

    @labels.setter #Setter that sets the labels
    def labels(self, labels):
        self._labels = labels
        self.notify() # Notify the observers whenever somebody changes the labels
        
    @property #Getter that gets the data
    def data(self):
        return self._data

    @data.setter #Setter that sets the labels
    def data(self, data):
        self._data = data
        self.notify() # Notify the observers whenever somebody changes the data
        

# This is the 'standard' view/observer which also acts as an 'abstract' class whereby deriving Bar/Chart/Table views override the update() method
# This 'abstracted' layer is always shown in examples but is important to demonstrate potential polymorphic behaviour of update()
class View():
    
    def __init__(self, name=""):
         self._name = name #Set the name of the Viewer

    def update(self, subject): #Alert method that is invoked when the notify() method in a concrete subject is invoked
        print("Generalised Viewer '{}' has: \nName = {}; \nLabels = {}; \nData = {}".format(self._name, subject._name, subject._labels, subject._data))

# Table 'chart' viewer/ConcreteObserver - overrides the update() method
class TableView(View):
    
    def update(self, subject): #Alert method that is invoked when the notify() method in a concrete subject is invoked
        fig = plt.figure(dpi=80)
        ax = fig.add_subplot(1,1,1)
        table_data = list(map(list,zip(subject._labels, subject._data)))
        table = ax.table(cellText=table_data, loc='center')
        table.set_fontsize(14)
        table.scale(1,4)
        ax.axis('off')
        plt.show()

# Bar chart viewer/ConcreteObserver - overrides the update() method
class BarView(View):
        
    def update(self, subject): #Alert method that is invoked when the notify() method in a concrete subject is invoked
        objects = subject._labels
        y_pos = np.arange(len(objects))
        performance = subject._data
        plt.bar(y_pos, performance, align='center', alpha=0.5)
        plt.xticks(y_pos, objects)
        plt.ylabel('Usage')
        plt.title('Programming language usage')
        plt.show()
        
# Pie chart viewer/ConcreteObserver - overrides the update() method
class PieView(View):

    def update(self, subject): #Alert method that is invoked when the notify() method in a concrete subject is invoked
        # Pie chart, where the slices will be ordered and plotted counter-clockwise:
        labels = subject._labels
        sizes = subject._data
        explode = (0.1, 0, 0, 0, 0, 0)  # only "explode" the 1st slice
        fig1, ax1 = plt.subplots()
        ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
        ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
        plt.show()

#Let's create our subjects
m1 = Model("Model 1")
m2 = Model("Model 2") # This is never used!

#Let's create our observers
v1 = View("1: standard text viewer")
v2 = TableView("2: table viewer")
v3 = BarView("3: bar chart viewer")
v4 = PieView("4: pie chart viewer")
####

#Let's attach our observers to the first model
m1.attach(v1)
m1.attach(v2)
m1.attach(v3)
m1.attach(v4)
####

# Let's just call the notify() method to see all the charts in their unchanged state
m1.notify()

# Now Let's change the properties of our first model
# Change 1 triggers all 4 views and updates their labels
m1.labels = ['C#','PHP','JavaScript','ASP','Python','Smalltalk']
# Change 2 triggers all 4 views and updates their data
m1.data = [1,18,8,60,3,1]
````

## <font color="red">Logbook Exercise 16: The Prototype Design Pattern</font> ##

### 1. Intent
The prototype design pattern allows for new objects to be made by cloning an existing one. It is a creational design pattern and allows the creation of these objects to use less resources and means that the client does not need to know the specific details of the class, allowing for the dynamic creation of these objects at runtime (Gamma et al., 1994).

### 2. Motivation
It is useful in scenarios where it is easier to make copy an object than to create it from scratch. This is because it bypasses the initialisation processes, which can be complicated for objects that have a lot of internal state that is not exposed externally. This allows the object to directly copy the internal state, thus reducing the steps and by extension, the amount of computational resources required to create the object (Refactoring Guru, n.d). This also makes the whole process simpler, allowing code to adhere to the KISS principle better.

The design pattern is useful in object-oriented programming languages like Java for polymorphism as objects can be created using just a clone method. This means that the object can be created at runtime without requiring detailed knowledge of the underlying class, allowing the code to be much more flexible (Cooper, 2000). For example, in the example of a photo editing application, we may want to copy a shape or an image. The prototype design pattern would allow us to create the shape or image as a prototype, and then clone it, allowing it to look exactly the same as the original. This means that we do not have to make all of the transformations we may have already made to it previously. As each transformation requires computational power, we can save processing resources. 

The prototype design pattern also benefits dynamically typed languages like Python as it enables flexibility without compromising the complexity and maintainability of the codebase. In dynamically typed languages, it allowsobjects to be modified and created at runtime which is important for scenarios where the structure of objects can vary a lot. For example, in a web app (which typically use the dynamically typed Javascript language), managing user sessions may use a session object that contain a unique set of attributes (Giridhar, 2016). This means that these are better implemented using the cloning of a prototype to allow for this flexibility. This adaptibility is consistent across the use for both statically types and dynamically typed languages, allowing the prototype design pattern to both simplify the creation of object and enabling more efficient use of computational resources. The Prototype pattern is used best when classes are defined at runtime or when a class may have only a few variations in its internal state. 

### 3. Structure

The Prototype pattern at its core includes a client, a prototype interface and concrete implementations of this interface. The prototype interface has a clone method, and the concrete implementations can have a clone method that overrides this, enabling polymorphism (Gamma et al., 1994). The interaction between these components are shown in Figure 16.1. There are two types of clone method that can be used, shallow and deep cloning . Shallow cloning creates a new instance of the class and copies the fields to the instance. This means that if the field is a reference type, then the underlying object referred in the clone is the same as that in the original. A deep clone on the other hand copies the objects for any reference fields as well such that the clone does not share any objects with the original (Refactoring Guru, n.d.). This means that is a completely independent object.



Figure 16.1 - UML diagram of the Prototype Design Pattern

For more complex systems, a prototype registry can also be used to simplify the management of many different prototypes. This registry can be used to store a set of pre-initialised prototypes representing different states. This allows the client to request clones from the registry, bypassing the need for it to know the details of the prototypes configuration.


### 4. Implementation
The prototype design pattern shows most utility within statically typed languages like Java as it is able to abstract the creating of objects. Statically typed languages usually require the details of an object's class to be known before it is able to create it. The prototype method provides a way of bypassing this necessity and polymorphism and encapsulation can be used to manipulate the objects by overriding the clone method provided by the interface. This is useful for situations where the class may vary (Budd, 2004).

The pattern also helps the system adhere to the open-closed principle as logic to create objects is encapsulated in the the prototype object (Martin, 2003). This makes the code more maintainable and reusable as it keeps internal state consistent between clone (Budd, 2004). It also has performance benefits as it removes the need for the instantiation processes to be carried out again for the second object, it can be cloned as it is (Shalloway & Trott, 2004). This means that objects that have complex constructor logic only need to run the constructor logic once. 

Whilst the benefit is greater for statically typed languages, dynamically typed languages such as Python are able to modify objects during runtime. This means that the Prototype design pattern can complement this by creating objects without needing to set the internal state again. This allows these objects to be created and reconfigured as needed (Stefanov, 2010).

The Prototype Pattern can vary in implementation between statically and dynamically typed languages. For example, in the statically typed Java, a Cloneable interface is built into the language which can be extended and the copy method within can be overridden. This is by default a shallow clone (Oracle, n.d.). This means that to implement deep cloning, this clone method needs to be overridden with a method that carefully copies each attribute completely so that they are independent from the originals in memory. Dynamically typed Python on the other hand uses the copy module, which provides the method to copy the object. It natively supports both shallow and deep copying. The module makes it simple to clone without needing to manually copy attributes. The deepcopy method makes it especially easy to create deep copies of objects as it automatically duplicates any nested objects (Python docs, n.d.).

### 5. Sample Code

````python
import copy

# Class to act as prototype interface and provide clone methods
class Prototype:
    def clone(self):
        return copy.copy(self)

    def deep_clone(self):
        return copy.deepcopy(self)

class Dimensions:
    def __init__(self, lengths, angles):
        self.lengths = lengths  # List of lengths
        self.angles = angles    # List of angles

class Shape(Prototype):
    def __init__(self, name, dimensions):
        self.name = name
        self.dimensions = dimensions

# Create shape and shallow copy it
original_shape = Shape("Triangle", Dimensions([3, 4, 5], [60, 60, 60]))
shallow_copied_shape = original_shape.clone()

print(f"Original lengths before shallow copy is modified: {original_shape.dimensions.lengths}")
print(f"Shallow lengths before shallow copy is modified: {shallow_copied_shape.dimensions.lengths}")
print(f"Original lengths are equal to shallow copy lengths: {original_shape.dimensions.lengths == shallow_copied_shape.dimensions.lengths}")
print()

# Modify lengths of shallow copy
shallow_copied_shape.dimensions.lengths.append(6)
    
print(f"Original lengths after shallow copy is modified: {original_shape.dimensions.lengths}")
print(f"Shallow lengths after shallow copy is modified: {shallow_copied_shape.dimensions.lengths}")
print("Shallow copy shares lengths with original:", shallow_copied_shape.dimensions.lengths == original_shape.dimensions.lengths)
print("Shallow copy lengths stored at same memory address as original:", id(shallow_copied_shape.dimensions.lengths) == id(original_shape.dimensions.lengths))
print()

# Deep copy the original shape
deep_copied_shape = original_shape.deep_clone()

print(f"Original lengths before deep copy is modified: {original_shape.dimensions.lengths}")
print(f"Shallow lengths before deep copy is modified: {deep_copied_shape.dimensions.lengths}")
print(f"Original lengths are equal to deep copy lengths: {original_shape.dimensions.lengths == deep_copied_shape.dimensions.lengths}")
print()

# Modify lengths of deep copy
deep_copied_shape.dimensions.lengths.append(7)

print(f"Original lengths after deep copy is modified: {original_shape.dimensions.lengths}")
print(f"Deep lengths after deep copy is modified: {deep_copied_shape.dimensions.lengths}")
print("Deep copy shares lengths with original:", deep_copied_shape.dimensions.lengths == original_shape.dimensions.lengths)
print("Deep copy lengths stored at same memory address as original:", id(deep_copied_shape.dimensions.lengths) == id(original_shape.dimensions.lengths))
````

### 6. Evaluation

The Prototype design pattern offers advantages and disadvantages across different languages and their respective paradigms. This section will evaluate the design pattern using Python to represent a dynamically typed language and Java to represent a statically typed language.

#### Programming Language Idioms

Being a dynamically typed language, Python naturally complements the flexibility of the Prototype design pattern as objects can be cloned and modified at runtime without the need of boilerplate code. This makes it more efficient and readable and Python's copy package makes it easy to implement. However, as there is so much built into the language in functionality for creating prototypes, the utility of the design pattern is less than when it is implemented in a statically typed language. JavaScript, another dynamically typed language, also has inheritance that naturally complements the design pattern (Crockford, 2008). Also, in these languages, prototypes can be created dynamically so a prototype registry becomes useful to implement to manage the various prototypes (Refactoring Guru, n.d.). The Prototype pattern is arguably more useful in statically typed languages like Java as it provides a way to create an object without the code being coupled to the object's class. Java is able to implement it by using the Cloneable interface and polymorphism to override the clone method where necessary.

#### Consequences
The benefits of the design pattern are that it allows objects to be created without complex instantiation processes (Freeman et al., 2004). This means that it often reduces computational cost. Use of a prototype registry can also make the pattern more useful but it also adds complexity to the code in terms of state management, so this is another important consideration in the design and implementation of this design pattern.

When using the design pattern, it is also important to distinguish between shallow and deep copying. Shallow copying can potentially lead to unexpected behaviour as changes to some of the attributes in the copied object can change that of the original (as shown in the code example above). Deep copying helps to mitigate this but needs to consider how this affects performance, especially in cases where objects are very complex. Deep copying is simple in Python thanks to the copy module but Java requires the clone method from the Cloneable interface to be overridden. This means that it is on the developer to ensure that all objects are copied properly (Bloch, 2008). This increases code complexity so developers need to be wary of bugs.

#### Examples of Appropriate uses
The prototype pattern is useful in software that requires large amounts of computational power. For example, videogames can use the Prototype pattern to copy sprites quickly. This helps them to improve performance and the reusability helps code abide by the DRY principle (McLaughlin & Pollice, 2016). Another good example of this is in photo editing software as image layers can also easily be cloned.

Another example is an analytics dashboard for a business. Here, users can customise widgets so that they can customise it to whatever suits their job role best. The Prototype pattern can be used to provide an Prototype interface called widget, with a clone method. All the possible widgets could then be stored in a WidgetRegistry, allowing them to be easily cloned. At this point, the user could add a new widget, and instead of running all of the complex code to create the widget object from scratch, it would just be cloned from the registry and the user could then set the dataset they would like to see. The user could then add a similar widget by cloning the widget and simply changing the dataset. This helps to improve the users experience as it allows for customisability in a resource efficient manner.

### Design Considerations
When compared to other creational design patterns like the Factory pattern, it is important to consider the use case. The Prototype design pattern is most useful in systems where similar complex objects are created often. On the other hand, another pattern like the Factory pattern can be more useful when the system needs to be more scalable or requires the addition of new types (Beck & Cunningham, 1987). In more complex applications, both patterns can also be used together in different parts of the application. For example, a word processor may want to use the Factory design pattern to create new document templates but the Prototype pattern for new documents using specific templates. To make this work, the Factory pattern could create the prototypes that would then be used for the Prototype pattern.

### 1. Intent

The command design pattern is a behavioural design pattern designed to "Encapsulate a request as an object, thereby letting you parameterize clients with different requests, queue or log requests, and support undoable operations" (Gamma et al., 1994). This means that all of the information required to carry out that request is stored within the request object or command itself. This allows it to be used in several parts of the application.

### 2. Motivation

The command design pattern allows us to model commands as objects so that we can apply object oriented programming principles to it. This is useful as it allows us to decouple commands from the objects that invoke it. This makes it easier to apply the single responsibility principle as the encapsulation of the command means that the command object is only responsible for doing one thing. This is extremely useful in the design of user interface toolkits as the toolkit designers are unable to know how specific applications will use the components in the toolkit, so they need to be made as generic as possible (Gamma et al., 1994).

The command design pattern also makes it easy to queue commands so that they happen in a certain order. This is extremely useful as a program may require a certain order of things to happen so that data is available when it needs to be (Object Design, n.d.). A good example of this is in an installation wizard. When installing certain programs, certain dependencies may need to be installed first so that the installation wizard can install the software properly. The command design pattern can be used in this scenario to queue the commands for the various dependencies so that they happen in order.


### 3. Structure



The command design pattern requires a command, a receiver, an invoker and the client. The command interface defines the execute method for executing a command. The client implements this interface to create a concrete command object which is able to use the execute method, and it sets it to a receiver. An invoker is the object that asks the command object to make a request by calling its execute method. Finally, a receiver is an object that carries out the work required for the command to be fulfilled. Additionally, it can also be used in distributed systems for task scheduling. The encapsulation that the design pattern provides allows the commands to be queued and prioritised, allowing the pattern to adapt to complex requirements. An example of a model to demonstrate this design pattern is the example of turning a light bulb on and off using a light switch. In this situation, the light bulb would be the receiver as it carries out the work of the command, it either comes on or off. The command interface is analogous to changing the state of the lightbulb, and the concrete commands would be turning the light bulb on and turning the light bulb off. The light switch acts as the invoker, and it doesn't know what is happening when its switched, it just knows that something is happening. Finally, the client is the wiring as this allows the command to be set to the receiver (light bulb). Figure 17.2 shows the Command design pattern being applied in this context.

### 4. Implementation

A command can range from being a simple action, with the only information it having being that it is carrying out a request to fully encapsulating everything required for the request. It is important to consider the use case for the design pattern. For example, for a UI toolkit, a component has no idea about the receiver of the command, so the command object would need to fully implement as much as possible in order for the component to be useful across multiple use cases.

The command design pattern also allows for undoing and redoing of commands, which can be incredibly useful in creative applications such as text editors. In order to implement these features, an undo command needs to be provided to reverse the command. Additional state would also need to be included to implement this as the application would need to know the last command applied. To support more than one undo and redo at a time, the application would need to build up a history of the commands being executed. A key consideration when designing systems with undo and redo functionality is that it is easy for errors to build  up if the commands do not contain enough information to allow for the state of the receiver to be restored. Maintaining a command history also has performance implications so this must be considered.

### 5. Sample Code

To demonstrate the command design pattern, a basic text editor will be created using the tkinter module. The requirements for this are defined below:

* The system will allow users to add text to a text editor through a graphical user interface.
* The system will allow users to undo the last text addition
* The system will allow users to queue multiple undo actions
* The system will be intuitive to use, with clear options for adding text and undoing actions
* The system will maintain a history of actions to allow for undoing and redoing

The code is shown in the cell below.

````python
# Python doesn't have built in interfaces but ABC can provide support for abstract methods

from abc import ABC, abstractmethod

# Tinkter imported to make a simple GUI
import tkinter as tk
from tkinter import simpledialog

# Command Interface
class Command(ABC):
    @abstractmethod
    def execute(self):
        pass

    @abstractmethod
    def undo(self):
        pass

# Text Editor Class
class TextEditor:
    def __init__(self, text_widget):
        self.text_widget = text_widget

    def add_text(self, text):
        self.text_widget.insert(tk.END, text)

class AddTextCommand(Command):
    def __init__(self, editor: TextEditor, text: str):
        self.editor = editor
        self.text = text
        self.previous_text = editor.text_widget.get("1.0", tk.END)

    def execute(self):
        self.editor.add_text(self.text)

    def undo(self):
        self.editor.text_widget.delete("1.0", tk.END)
        self.editor.text_widget.insert(tk.END, self.previous_text)
        

class CommandInvoker:
    # Invoker needs to store a history to enable commands to be undone
    def __init__(self):
        self.history = []

    # Executes command whilst adding it to the history
    def execute_command(self, command):
        self.history.append(command)
        command.execute()

    # Removes the last command from the history and calls its undo command
    def undo(self):
        if self.history:
            command = self.history.pop()
            command.undo()

# Application class
class EditorApp:
    def __init__(self, root):
        self.root = root
        self.text_editor_widget = tk.Text(root, height=10, width=40)
        self.text_editor_widget.pack()
        
        # Creating the editor and command invoker 
        self.editor = TextEditor(self.text_editor_widget)
        self.invoker = CommandInvoker()

        # Making buttons required to add text and undo the last command
        add_text_button = tk.Button(root, text="Add Text", command=self.add_text)
        add_text_button.pack()

        undo_button = tk.Button(root, text="Undo", command=self.undo)
        undo_button.pack()

    # Function to allow user to type some text and call the invoker to add the text
    def add_text(self):
        text = simpledialog.askstring("Input", "Enter text:")
        if text:
            self.invoker.execute_command(AddTextCommand(self.editor, text))

    # Function call the invoker to undo the last command    
    def undo(self):
        self.invoker.undo()

if __name__ == "__main__":
    root = tk.Tk()
    app = EditorApp(root)
    root.mainloop()
````

**Note:** In the sample code, if text is typed directly into the text field, the user is unable to undo it with the undo button. This was intentionally left in to demonstrate the importance of fully applying the design pattern. This occurs because the text was not added with a command that is in the command history, so it has no way of undoing it.

### 6. Evaluation
The Command Design Pattern is an extremely useful design pattern but its strengths and weaknesses should be evaluated based on the programming language it is implemented in and the requirements for the software.

#### Programming Language Idioms
Being a dynamically typed language, Python allows us to implement the design pattern easily. However, the lack of interfaces within the language means that we have to use the Abstract Base Classes module to simulate the Command class as an interface. This shows that Python is flexible as it contains modules that allow us to do this but as it is dynamically typed, it will not strictly enforce type checks in the same way as a statically typed language like Java might do (Martin, 2003). This could potentially lead to run time errors so the code needs to be rigorously tested to avoid this.

#### Consequences
The Command Design Pattern improves code maintainability by creating a clear separation of concerns (Fowler, 2002) by abstracting the invoking of a command from its execution. This helps to improve flexibility in the applications design, but it does so by introducing additional complexity. This is because there are extra classes and objects (Kerievsky, 2004). This means that the impact on performance is a key consideration, especially in more complex applications.

#### Examples of Appropriate Uses:
The Command Design Pattern is most appropriately used in scenarios where actions need to be abstracted away. A good example of this is in a graphical user interface, clicking a button will trigger a specific command. In this scenario, the command design pattern helps to simplify event handling and promotes code reusability (Gamma et al., 1994). This helps code adhere to the DRY principle. A real-world application for this is Adobe Photoshop, where actions like applying a filter is encapsulated as a command. This allows for undoing and redoing of the adjustment, which is of utmost importance within photo editing.

The Command Design Pattern's ability to implement undo and redo features also help improve the user's experience. The design pattern makes it easy to implement as the abstraction of command objects means that it is easy to store a history of commands executed. As the command object in this situation is responsible for both executing the command and reversing it, it makes it easy for the applications to go back or forwards through user actions. This feature is extremely useful for creative applications such as text editors, graphic design software and integrated development environments (Freeman et al., 2004).

#### Design considerations:
When designing applications with the Command Design Pattern in mind, the requirements and constraints of the application should be considered. If the application is designed for a system where performance is a key concern, the benefits of applying the design pattern need to be weighed up against the extra complexity that creating and managing command objects brings about (Martin, 2003). Additionally, in simpler applications, the abstraction that the design pattern requires may not be justified as it could be sufficient to execute commands directly.

In the example of a graphical user interface, the Strategy design pattern is also able to provide modularity and flexibility. When comparing against the Command pattern, each has their own advantages and disadvantages. For example, in a text editor, the encapsulation of actions that the Command pattern provides allows for specific actions to map to different keys or shortcuts (like Ctrl+C to copy) (Gamma et al., 1994). It also provides the way for undo and redo functionality to be implemented, which are essential to these types of applications. The Strategy pattern on the other hand is more suitable for situations where algorithms need to change quickly without changes to the client (Thompson & Rae, 2018). The best example of this is where the sorting of a list needs to be able to be sorted using different algorithms, but the end list will always be a list with the same items, just sorted differently. 

The Command pattern is best used where action management and the keeping track of command history is important and the strategy pattern is best when different contexts change the functionality of how the specific action is implemented (Morris, 2017). Therefore, it can be wise for complex applications like graphic design software to use a combination of both patterns where suitable to enable the application to get the best of both worlds. In a graphics design software, each modification to an image could be encapsulated as a command using the design pattern. For example, "Apply Filter" could be a command. This would allow it to be added to the command history, enabling it to be undone and redone. There are however many different types of filter that could be applied. This is where the strategy design pattern could come in as it would enable for different types of filters to be applied dynamically. So the command could be applied either using the "Sepia Filter" strategy or the "Monochrome" Filter strategy. This is demonstrated in the code block below. In the code, the ImageCommand class is used to encapsulate the filter and the image. The invoker is the Image editor class as it is responsible for executing the command. The Strategies act as the receivers as they are responsible for doing the work. Lambda functions enable the strategy to be passed in without having them execute within the invoker.

````python
# For image processing
import cv2
import numpy as np
# To make a GUI
from tkinter import *
from tkinter import filedialog
# To allow tkinter to use image files
from PIL import Image, ImageTk
# To allow Python to interact with system
import os
# To help deal with temporary files
import tempfile

# Abstract Strategy Interface
# Note that in this example, the strategies act as the receivers as they are responsible for carrying out the work
class FilterStrategy:
    def apply(self, image):
        raise NotImplementedError

# Concrete Strategy for Sepia filter
class SepiaFilterStrategy(FilterStrategy):
    def apply(self, image):
        # Sepia effect matrix
        kernel = np.array([[0.272, 0.534, 0.131],
                           [0.349, 0.686, 0.168],
                           [0.393, 0.769, 0.189]])
        # Applies matrix effect
        return cv2.transform(image, kernel)

# Concrete Strategy for Monochrome filter
class MonochromeFilterStrategy(FilterStrategy):
    def apply(self, image):
        # Converts image to greyscale
        return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Acts as Command Interface to enable multiple commands to be written out.
class ImageCommand:
    def __init__(self, strategy, image):
        self.strategy = strategy  # The filter strategy to use
        self.image = image
    
    def execute(self):
        # Execute the chosen strategy on the image.
        return self.strategy.apply(self.image)

# Invoker Class
class ImageEditor:
    def __init__(self, root):
        self.root = root
        self.root.title("Image Editor")
        self.image_label = Label(self.root)
        self.image_label.pack()

        self.command_history = []  # History of applied filters
        self.undo_history = []  # History of undone filters
        # History of undone filters is required to enable redoing

        # GUI elements for user interaction
        Button(self.root, text="Open Image", command=self.open_image).pack(side=LEFT)
        # Lambdas to pass strategy without executing
        Button(self.root, text="Apply Sepia", command=lambda: self.apply_filter(SepiaFilterStrategy())).pack(side=LEFT)
        Button(self.root, text="Apply Monochrome", command=lambda: self.apply_filter(MonochromeFilterStrategy())).pack(side=LEFT)
        Button(self.root, text="Undo", command=self.undo).pack(side=LEFT)
        Button(self.root, text="Redo", command=self.redo).pack(side=LEFT)

    # Open image file
    def open_image(self):
        file_path = filedialog.askopenfilename() 
        if file_path:
            self.image_path = file_path
            self.display_image(file_path)  
            self.original_image = cv2.imread(file_path)  
            self.temp_files = [self.image_path]
            # Clear history when new image is applied
            self.command_history.clear()
            self.undo_history.clear()

    # Apply filter strategy to image
    def apply_filter(self, strategy):
        if hasattr(self, 'original_image'):
            command = ImageCommand(strategy, self.original_image.copy())
            processed_image = command.execute()
            # Save the processed image as a temporary file
            temp_path = tempfile.mktemp(suffix='.jpg')
            cv2.imwrite(temp_path, processed_image)
            self.temp_files.append(temp_path)
            # Update histories
            self.command_history.append(temp_path)
            self.undo_history.clear()
            self.display_image(temp_path)

    # Undo
    def undo(self):
        if self.command_history:
            self.undo_history.append(self.command_history.pop())
            image_path = self.command_history[-1] if self.command_history else self.temp_files[0]
            self.display_image(image_path)

    # Redo
    def redo(self):
        if self.undo_history:
            redo_path = self.undo_history.pop()
            self.command_history.append(redo_path)
            self.display_image(redo_path)

    # Display image in tkinter
    def display_image(self, path):
        pil_image = Image.open(path)
        img = ImageTk.PhotoImage(pil_image)
        self.image_label.configure(image=img)
        self.image_label.image = img
    
    # Clean up
    def cleanup_temp_files(self):
        for file in self.temp_files[1:]:
            os.remove(file)

# Run application
if __name__ == "__main__":
    root = Tk()
    editor = ImageEditor(root)
    
    # Cleanup when window closes
    def on_closing():
        editor.cleanup_temp_files()
        root.destroy()

    root.protocol("WM_DELETE_WINDOW", on_closing)
    root.mainloop()
````
